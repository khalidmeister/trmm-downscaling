```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_presipitasi.tsv', sep='\t')
df.head()
```

```python
df.RR.isnull().value_counts()
```

```python
df['RR'][df.RR == 8888] = 0
df['RR'][df.RR == 9999] = 0
df['RR'][df.RR.isnull() == True] = 0

df.head()
```

```python
df.Tanggal = pd.to_datetime(df.Tanggal)
df.info()
```
